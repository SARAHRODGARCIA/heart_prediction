PADRONIZA√á√ÉO DOS DADOS:

Num√©ricas cont√≠nuas: age, height, weight ‚Üí recomend√°vel padronizar (StandardScaler ou MinMaxScaler).

Categ√≥ricas codificadas como n√∫meros (gender, cholesterol, gluc, smoke, alco, active)

gender (1/2) ‚Üí pode ser mantido como est√°.

cholesterol e gluc (1,2,3) ‚Üí Label Enconder.

smoke, alco, active (0/1) ‚Üí n√£o precisam de padroniza√ß√£o.

----------------------- O que ser√° feito:
Num√©ricas cont√≠nuas (age, height, weight) ‚Üí MinMaxScaler (escala entre 0 e 1)

Categ√≥ricas ordinais (cholesterol, gluc) ‚Üí LabelEnconder.

Bin√°rias (gender, smoke, alco, active) ‚Üí mantidas como est√£o

```python
from sklearn.preprocessing import MinMaxScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
import pandas as pd

# Colunas
num_cols = ['age', 'height', 'weight']
ord_cols = ['cholesterol', 'gluc']  # OrdinalEncoder
bin_cols = ['gender', 'smoke', 'alco', 'active', 'cardio_disease']  # mantidas

# Pr√©-processamento
preprocessor = ColumnTransformer(
    transformers=[
        ('num', MinMaxScaler(), num_cols),
        ('ord', OrdinalEncoder(categories='auto'), ord_cols)  # Label/Ordinal Encoder
    ],
    remainder='passthrough'  # mant√©m as colunas bin√°rias
)

# Aplicar transforma√ß√£o
X_scaled = preprocessor.fit_transform(base1[num_cols + ord_cols + bin_cols])

# Nomes das colunas transformadas
all_cols = num_cols + ord_cols + bin_cols  # agora ordinais em vez de One-Hot

# Criar DataFrame final base2 com todas as colunas padronizadas/substitu√≠das
base2 = pd.DataFrame(X_scaled, columns=all_cols)

# Conferir
print(base2.head())
print(base2.info())

```

# 7) EXPLICA√á√ÉO:


A) Regress√£o Log√≠stica:

A regress√£o log√≠stica √© um modelo usado para prever resultados que t√™m apenas duas op√ß√µes, como ‚Äútem doen√ßa card√≠aca‚Äù ou ‚Äún√£o tem‚Äù.
Calcula a probabilidade de um evento acontecer usando uma fun√ß√£o matem√°tica chamada sigmoide, que garante que o resultado fique sempre entre 0 e 1. Depois, a gente escolhe um limite (geralmente 0,5) para decidir a classe final.

Em outras palavras: ela transforma n√∫meros em chances e decide se a pessoa pertence a uma categoria ou outra.

B) Por que √© um modelo de classifica√ß√£o:

Porque o objetivo n√£o √© prever n√∫meros cont√≠nuos, mas sim decidir a qual grupo BINARIO algo pertence.
